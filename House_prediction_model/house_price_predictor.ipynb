````python
#convert range of total_sqft to avg of range
def range(x):
  temp = x.split('-')
  temp1= x.split('*')
  if len(temp)==2:
    return (float(temp[0])+float(temp[1]))/2
  if len(temp1)==2:
    return float(temp1[0])*float(temp1[1])
  try:
    return float(x)
  except:
    None

````

````python
house['total_sqft']=house['total_sqft'].apply(range)
house.head()
````

# model building


````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score

````

````python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,test_size=0.2)
````

````python
#applying linear regression
````

````python
column_trans =make_column_transformer((OneHotEncoder(sparse_output=False),['location']),remainder='passthrough')
````

````python
scaler = StandardScaler()
````

````python
reg = LinearRegression()
````

````python
pipe = make_pipeline(column_trans,scaler,reg)
````

````python
pipe.fit(X_train,Y_train)
````

````python
Y_pred_lr = pipe.predict(X_test)
````

````python
r2_score(Y_test,Y_pred_lr)
````

# Applying lasso

````python
lasso= Lasso()
````

````python
pipe = make_pipeline(column_trans,scaler,lasso)
````

````python
pipe.fit(X_train,Y_train)
````

````python
Y_pred_lasso=pipe.predict(X_test)
r2_score(Y_test,Y_pred_lasso)
````
